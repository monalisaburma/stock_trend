```python
# Count the number of entries per year
AAPL_data['Year'] = AAPL_data['Date'].dt.year
for i in range(2004, 2025):
    print(i, AAPL_data['Year'].tolist().count(i))
```

```python
# Calculate moving averages
AAPL_data['MA_for_100_days'] = AAPL_data['Adj Close'].rolling(100).mean()
AAPL_data['MA_for_200_days'] = AAPL_data['Adj Close'].rolling(200).mean()
AAPL_data['MA_for_250_days'] = AAPL_data['Adj Close'].rolling(250).mean()

# Filter out rows with NaN values in moving averages
AAPL_data_filtered = AAPL_data.dropna(subset=['MA_for_100_days', 'MA_for_200_days', 'MA_for_250_days'])

# Verify the filtered DataFrame
AAPL_data_filtered.head()
```

```python
# Plotting function with explicit labels
def plot_graph(figsize, values, labels, title):
    plt.figure(figsize=figsize)
    for value, label in zip(values, labels):
        plt.plot(value, label=label)
    plt.xlabel("Years")
    plt.ylabel("Price")
    plt.title(title)
    plt.legend()
    plt.show()

plot_graph((15, 5), [AAPL_data_filtered.set_index('Date')['MA_for_250_days']], ['MA for 250 Days'], 'MA for 250 Days of Google data')

# Plot MA for 250 days with Adj Close
plot_graph((15, 5), [AAPL_data_filtered.set_index('Date')['Adj Close'], AAPL_data_filtered.set_index('Date')['MA_for_250_days']],
           ['Adj Close', 'MA for 250 Days'], 'MA for 250 Days of AAPL Data')

# Plot MA for 200 days with Adj Close
plot_graph((15, 5), [AAPL_data_filtered.set_index('Date')['Adj Close'], AAPL_data_filtered.set_index('Date')['MA_for_200_days']],
           ['Adj Close', 'MA for 200 Days'], 'MA for 200 Days of AAPL Data')

# Plot MA for 100 days with Adj Close
plot_graph((15, 5), [AAPL_data_filtered.set_index('Date')['Adj Close'], AAPL_data_filtered.set_index('Date')['MA_for_100_days']],
           ['Adj Close', 'MA for 100 Days'], 'MA for 100 Days of AAPL Data')

# Plot the whole data including all relevant columns
plot_graph((15, 5), [AAPL_data_filtered.set_index('Date')['Adj Close'], AAPL_data_filtered.set_index('Date')['MA_for_100_days'], AAPL_data_filtered.set_index('Date')['MA_for_250_days']],
           ['Adj Close', 'MA for 100 Days', 'MA for 250 Days'], 'Adjusted Close and Moving Averages')
```

```python
Adj_close_price = AAPL_data_filtered[['Adj Close']]
```

```python
max(Adj_close_price.values),min(Adj_close_price.values) 
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(Adj_close_price)
scaled_data
```

```python
len(scaled_data)
```

```python
# Prepare training and testing datasets
x_data = []
y_data = []
for i in range(100, len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])

import numpy as np
x_data, y_data = np.array(x_data), np.array(y_data)
```

```python
x_data[0],y_data[0]
```

```python

int(len(x_data)*0.7)
```

```python
4908-100-int(len(x_data)*0.7)
```

```python
splitting_len = int(len(x_data)*0.7)
x_train = x_data[:splitting_len]
y_train = y_data[:splitting_len]

x_test = x_data[splitting_len:]
y_test = y_data[splitting_len:]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
# Build and train the LSTM model
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, batch_size=1, epochs=2)
```

```python
model.summary()
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
inv_predictions = scaler.inverse_transform(predictions)
inv_predictions
```

```python
inv_y_test = scaler.inverse_transform(y_test)
inv_y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# Calculate evaluation metrics
mae = mean_absolute_error(inv_y_test, inv_predictions)
mse = mean_squared_error(inv_y_test, inv_predictions)
rmse = np.sqrt(mse)
r2 = r2_score(inv_y_test, inv_predictions)

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'R^2: {r2}')
```

```python
# Create a DataFrame for plotting predicted vs actual values
ploting_data = pd.DataFrame({
    'Date': AAPL_data_filtered['Date'][splitting_len + 100:].values,
    'original_test_data': inv_y_test.reshape(-1),
    'predictions': inv_predictions.reshape(-1)
})

ploting_data.head()
```

```python
ploting_data.set_index('Date', inplace=True)
```

```python
ploting_data.head()
```

```python
# Plot the test data with predictions with custom labels
plt.figure(figsize=(15, 6))
plt.plot(ploting_data['original_test_data'], label='Original Test Data')
plt.plot(ploting_data['predictions'], label='Predicted Test Data')
plt.xlabel("Years")
plt.ylabel("Test Data with Predictions")
plt.title("Test Data with Predictions")
plt.legend()
plt.show()

```

```python
# Ensure the indices are properly set for concatenation
Adj_close_price = Adj_close_price.set_index(AAPL_data_filtered['Date'])
```

```python
# Plot the whole data including predictions with custom labels
whole_data = pd.concat([Adj_close_price[:splitting_len + 100], ploting_data[['original_test_data', 'predictions']]], axis=0)
whole_data.index = pd.to_datetime(whole_data.index)  # Ensure the index is datetime

plt.figure(figsize=(15, 6))
plt.plot(whole_data['Adj Close'], label='Training Data')
plt.plot(whole_data['original_test_data'], label='Original Test Data')
plt.plot(whole_data['predictions'], label='Predicted Test Data')
plt.xlabel("Years")
plt.ylabel("Whole Data")
plt.title("Whole Data of AAPL data")
plt.legend()
plt.show()
```
